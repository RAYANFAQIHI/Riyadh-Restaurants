# **Detecting & Dealing with Missing Values**
Detecting Missing Values Using isna()

```python
df.info()
```

```python
df.isna().sum()
```

```python
sns.heatmap(df.isna())
```

```python
(df.isna().mean()*100).sort_values(ascending=False)
```

```python
df["price"] = df["price"].astype(str).str.strip().str.title()
```

```python
df
```

```python
for c in ["likes","photos","tips","rating","ratingSignals"]:
    df[c] = pd.to_numeric(df[c], errors="coerce")
```

```python
df
```

# **Removing Missing Data Using dropna()**

```python
df_rated = df.dropna(subset=["rating", "ratingSignals"]).copy()
df_rated.shape
```

```python
df["likes"] = df["likes"].fillna(0)
```

```python
df["price"] = df["price"].fillna("Unknown")
df["price"].value_counts()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.duplicated(subset=["name","address"]).sum()
```

```python
df = (df.sort_values(["ratingSignals","rating","likes"], ascending=False)
        .drop_duplicates(subset=["name","address"], keep="first"))
```

# **Text Cleaning**

```python
for c in ["name","categories","address"]:
    df[c] = df[c].astype(str).str.strip()
```

# **Insight**

**The most common types (e.g. cafe/coffee shop)**

```python
top_cats = df["categories"].value_counts().head(15)
top_cats
```

```python
top_cats.plot(kind="bar")
plt.title("Top 15 Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

**Rating Distribution (Rated only)**

```python
df_rated["rating"].describe()
```

```python
df_rated["rating"].hist(bins=20)
plt.title("Rating Distribution (Rated Places Only)")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

**Avg Rating by Price (Rated only)**

```python
avg_rating_price = df_rated.groupby("price")["rating"].mean().sort_values(ascending=False)
avg_rating_price
```

```python
avg_rating_price.plot(kind="bar")
plt.title("Average Rating by Price (Rated Only)")
plt.xlabel("Price")
plt.ylabel("Avg Rating")
plt.tight_layout()
plt.show()
```

**Correlation (Engagement vs Rating)**

```python
plt.scatter(df_rated["ratingSignals"], df_rated["rating"])
plt.title("Rating vs Rating Signals")
plt.xlabel("ratingSignals")
plt.ylabel("rating")
plt.tight_layout()
plt.show()
```

```python
df["has_rating"] = df["rating"].notna().astype(int)
```

```python
df["signals_band"] = pd.cut(
    df["ratingSignals"],
    bins=[-1, 0, 10, 30, 100, 500, 1_000_000],
    labels=["0", "1-10", "11-30", "31-100", "101-500", "500+"]
)
```

```python
df["engagement"] = df["likes"].fillna(0) + df["tips"].fillna(0) + df["photos"].fillna(0)
```

```python
df = df.drop(columns=["lat","lng"])
```
